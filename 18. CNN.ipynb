# 18.1 전통적인 방식

머신러닝과 딥러닝의 가장 큰 차이중 하나가 손수 만든 특징의 활용 여부라고 했다.  
만약에 사진에서 어떤 물체를 인식하는 데 있어 경계를 추출하는 것이 중요하다고 했을 때, 미리 정해진 합성곱 커널(또는 필터) 을 사용하면 원하는 형태의 패턴 추출이 가능했다.  
즉, 합성곱 커널의 종류에 따라 추출할 수 있는 특징의 패턴은 바뀌기 떄문에 미리 세운 가설에 따라 커널을 설계하고 적용하는 형태로 프로세스가 진행되었다.

# 18.3 패턴 추출의 원리

전통적인 방식의 머신러닝에서는 가정을 통해 미리 설정한 합성곱 커널을 사용해서 패턴을 추출한다. 하지만 딥러닝에서는 모델의 출력 y^과 타깃값 y가 같아지도록 패턴을 추출할 수 있는 커널이 자동으로 학습된다.  
만약 모델이 특징을 추출하기 위해 양 대각선 및 가로 직선 방향의 패턴을 위한 커널을 사용해야 손실 값이 낮아진다면 경사하강법에 의해서 자동으로 패턴을 위한 커널을 찾도록 학습될 것이다.  

즉, 전통적인 방식에서는 손수 만든 특징을 바탕으로 합성곱 연산을 활용했다. 하지만 딥러닝에서는 합성곱 계층(CNN)을 활용하여 최적의 패턴 추출 방법을 자동으로 학습하고, 이렇게 학습된 커널을 사용하여 더 나은 패턴 추출을 수행한다.

### 1. 채널

여러 가지 특징의 패턴을 동시에 추출하고 싶을 때 여러개의 커널을 적용할 수 있다. 이렇게 사용되는 커널의 숫자를 출력 채널의 숫자라고 부르기도 한다.  
또한 특징을 추출하고자 하는 입력이 어러 개의 채널을 가질 수도 있는데 이것을 입력 채널이라고 부른다.

### 3. 합성곱 계층의 특징과 장단점

MNIST같은 이미지 샘플을 오른쪽으로 한 칸씩 이동시키더라도 샘플의 의미는 전혀 변하지 않으며 모델이 숫자를 인식하는 경우에도 전혀 차질이 없어야 한다.  
하지만 선형 계층은 태생부터 이와 같은 이미지의 변화를 받아들이기 힘든 구조이며, 모델의 입장에서 한 칸 이동한 그림은 전혀 다른 새로운 샘플이 된다.  
즉, 다시 새롭게 학습해야하는 대상이 된다.  
하지만 합성곱 계층은 위치에 상관없이 특징의 패턴을 인식할 수 있기 때문에 특정 부분이 한 칸 이동한 그림도 어려움 없이 해결이 가능하다.

또한 선형 계층의 경우 입력과 출력 벡터의 크기가 정해지면 내부 가중치 파라미터의 크기가 정해지는데 반해, 합성곱 계층은 입출력 크기와 상관없이 내부 커널의 크기에 따라 가중치 파라미터의 크기가 정해지므로 훨씬 적은 파라미터 수를 가지게 된다.  
모델이 더 적은 파라미터수를 가지게 되면 속도나 메모리 측면에서도 이득이고, 최적화에 있어서도 난이도가 낮아진다.  
선형계층에 비해서 병렬 연산으로 구성하기 쉬운 구조이므로 학습과 추론 면에서도 훨씬 월등하다.  

하지만 합성곱 계층은 입력과 출력 크기가 고정되어 있지 않다는 특징이 단점으로 작용할 수 있다.  
합성곱 계층은 입출력 크기를 직관적으로 확인하기 어렵기 때문에 합성곱 계층을 활용한 전체 신경망 구현의 난이도가 높아지게 된다.

# 18.4 맥스풀링과 스트라이드 기법

보통 심층신경망은 고차원 공간의 데이터를 저차원 공간에 비선형적으로 풀어낸 표현 과정을 통해 샘플의 특징을 추출하고 학습한다.  
하지만 합성곱 계층의 3x3 kernel + 1 Padding 형태는 입출력 크기가 같기 때문에 차원 축소 과정을 거친다고 보기 어렵다.  
이때 맥스 풀링 또는 스트라이드 기법을 활용한다.

### 2. 스트라이드

스트라이드는 커널이 기존 방식과 다르게 성큼성큼 칸을 건너뛰어 동작하는 것이다.

기존과 다르게 가로 세로 한 칸씩을 건너뛰며 합성곱 연산이 수행되었기 때문에, 출력 행렬의 크기가 2x2이다.

스트라이드는 코드로 구현할 때 보통 합성곱 계층의 설정값으로 주어 사용한다.  
따라서 도 다른 계층이 추가되는 맥스 풀링 기법과 달리 합성곱 계층 하나로 원하는 작업을 수행할 수 있다는 장점이 있다.  
하지만 스트라이드 기법은 합성곱 연산 자체를 건너뛰어 버리기 때문에 큰 값을 놓칠 수도 있다는 단점이 존재한다.
