# 18.1 전통적인 방식

머신러닝과 딥러닝의 가장 큰 차이중 하나가 손수 만든 특징의 활용 여부라고 했다.  
만약에 사진에서 어떤 물체를 인식하는 데 있어 경계를 추출하는 것이 중요하다고 했을 때, 미리 정해진 합성곱 커널(또는 필터) 을 사용하면 원하는 형태의 패턴 추출이 가능했다.  
즉, 합성곱 커널의 종류에 따라 추출할 수 있는 특징의 패턴은 바뀌기 떄문에 미리 세운 가설에 따라 커널을 설계하고 적용하는 형태로 프로세스가 진행되었다.

# 18.3 패턴 추출의 원리

전통적인 방식의 머신러닝에서는 가정을 통해 미리 설정한 합성곱 커널을 사용해서 패턴을 추출한다. 하지만 딥러닝에서는 모델의 출력 y^과 타깃값 y가 같아지도록 패턴을 추출할 수 있는 커널이 자동으로 학습된다.  
만약 모델이 특징을 추출하기 위해 양 대각선 및 가로 직선 방향의 패턴을 위한 커널을 사용해야 손실 값이 낮아진다면 경사하강법에 의해서 자동으로 패턴을 위한 커널을 찾도록 학습될 것이다.  

즉, 전통적인 방식에서는 손수 만든 특징을 바탕으로 합성곱 연산을 활용했다. 하지만 딥러닝에서는 합성곱 계층(CNN)을 활용하여 최적의 패턴 추출 방법을 자동으로 학습하고, 이렇게 학습된 커널을 사용하여 더 나은 패턴 추출을 수행한다.

### 1. 채널

여러 가지 특징의 패턴을 동시에 추출하고 싶을 때 여러개의 커널을 적용할 수 있다. 이렇게 사용되는 커널의 숫자를 출력 채널의 숫자라고 부르기도 한다.  
또한 특징을 추출하고자 하는 입력이 어러 개의 채널을 가질 수도 있는데 이것을 입력 채널이라고 부른다.

<img src = "cnn4.jpg" width = "400" height = "200">

위 그림처럼 커널의 개수가 최종 출력물의 채널 개수가 된다. 이때 각 커널은 입력 채널에 대해서 합성곱 연산을 수행하고 모두 더해서 출력 채널에 값을 표현한다.
